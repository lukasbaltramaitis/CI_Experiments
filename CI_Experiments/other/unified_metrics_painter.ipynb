```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
path = "/Users/lukas/Desktop/CI_Experiments/CI_Experiments/results/BPIC_17/2023-04-14_21-30-31"
```

```python
ACTIVITIES = 'activities'
ACTIVITIES_NUMERIC = 'activities_numeric'
ACTIVITIES_NUMERIC_POS_NEG = 'activities_numeric_pos_neg'
ACTIVITIES_POS_NEG = 'activities_pos_neg'
QINI = 'qini'
CUMLIFT = 'cumlift'
GAIN = 'gain'
CSV = 'csv'
```

```python
def plot_all(path, type_):
    approaches = [ACTIVITIES, ACTIVITIES_POS_NEG, ACTIVITIES_NUMERIC, ACTIVITIES_NUMERIC_POS_NEG]
    data = None
    for approach in approaches:
        try:
            full_path = f"{path}/{approach}/{type_}.{CSV}"
            data = pd.read_csv(full_path)
            plt.plot(data[data.columns[0]], data[data.columns[1]], label=approach)
        except FileNotFoundError:
            pass
    if type_ is QINI or type_ is GAIN:
        plt.plot((0, data[data.columns[0]].iat[-1]), (0, data[data.columns[1]].iat[-1]), label='random')
    plt.ylabel(type_)
    plt.xlabel('population')
    plt.legend()
    plt.show()
```

```python
plot_all(path, GAIN)
```
