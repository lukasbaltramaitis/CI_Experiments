```python
import pandas as pd
import matplotlib.pyplot as plt
from ylearn import Why
```

```python
path = "/Users/lukas/Desktop/CI_Experiments/CI_Experiments/data/prepared_process_logs/BPIC_17/activities/train.csv"
test_path = "/Users/lukas/Desktop/CI_Experiments/CI_Experiments/data/prepared_process_logs/BPIC_17/activities/test.csv"
```

```python
data = pd.read_csv(path)
data.head()
```

```python
why = Why()
r = why.identify(data, 'Outcome')
```

```python
r
```

```python
why.fit(data, 'Outcome', treatment=['Activity_A_Accepted', 'Activity_A_Cancelled', 'Activity_A_Complete'])
```

```python
why.causal_effect()
```

```python
test_data = pd.read_csv(test_path)
test_data.head()
```

```python
uplift = why.uplift_model(test_data, treatment='Activity_A_Accepted')
```

```python
qini = uplift.get_qini()
top = qini[qini.columns[0]].iat[-1]
idx = qini.index[-1]
print(idx)
top
```

```python
plt.plot(qini.index, qini[qini.columns[0]], label='Activity_A_Accepted')
plt.plot((0, idx), (0, top), label='random')
plt.xlabel('Population')
plt.ylabel('Qini')
plt.legend()
plt.show()
```
